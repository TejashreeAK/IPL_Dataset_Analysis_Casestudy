```python
#drop umpire3 column as it contains more than 75% of null values
df.drop("umpire3",axis=1,inplace=True)
```

```python
df.head()
```

```python
#drop null values of city , winner ,player_of_match,umpire1,umpire2 column
df.dropna(subset=["city","winner","player_of_match","umpire1","umpire2"])
```

```python
#which city hosted most number of matches?
city_matches=df.groupby(["city"])[["id"]].count().sort_values("id",ascending=False)[:1]
city_matches
```

```python
#draw bar plot and write down your insights
matches=df.groupby(["city"])[["id"]].count().sort_values("id",ascending=False)
plt.figure(figsize=(15,10))
matches.plot.bar()
plt.xlabel("City")
plt.ylabel("No. Of Matches")
plt.title("No. of Matches Played in Each City")
plt.show()
```

```python
#now compare in which venue of mumbai most number of matches played (draw bar plot and  write down insights)
mumbai=df[df["city"]=="Mumbai"].venue
b=df.groupby(mumbai)[["id"]].count().sort_values("id",ascending=False)
plt.figure(figsize=(20,15))
b.plot.bar()
plt.xlabel("Venue in Mumbai")
plt.ylabel("No. Of Matches")
plt.title("Venues of mumbai most number of matches played")
plt.xticks(rotation=45)
plt.show()
```
